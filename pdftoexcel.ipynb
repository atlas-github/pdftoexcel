#Import package and read pdf

```
import tabula

# Read pdf into list of DataFrame
sample_list = tabula.read_pdf("cimb-2019-financial-statements.pdf", pages='5')
sample_list
```

#Extract data for dataframe

```
df = sample_list[0]
df
```

#Clean dataframe

```
#clean header row

df = df.rename(columns = {"Unnamed: 0" : "Key Highlights"})
df1 = df.drop([0])

#merge rows 5 & 6

df1 = df1.replace(to_replace = "financing", value = "Expected credit losses on loans, advances and financing")
df2 = df1.drop([5])

#merge rows 35 & 36

df2 = df2.replace(to_replace = "– basic", value = "Earnings per share (sen)")
df3 = df2.drop([35])

#replace NaNs
df3 = df3.fillna('')
df3
```
